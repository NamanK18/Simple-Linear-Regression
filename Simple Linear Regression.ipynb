# SIMPLE LINEAR REGRESSION(SLR)

### This is an algorithm which is used for prediction.It has two types of variables:Independent and dependent variables.
### We are trying to predict dependent variable based on one independent variable.If the number of independent variableare more than one then it is known as MLR or Multiple Linear Regression.So the basic difference between the two techniques is the number of independent variables.
## If it is one then SLR ,If greater than one then MLR.
## The name of the dataset is Hotel.csv and is imported as a dataframe in python.
### The dependent variable is that which you want to predict.In our case it is revenue(for hotel file).The independent variable is the one which you are using to predict the dependent variable.In our case it is occupancy. Create a linear regression model to calculate the dependent variable(dv) and occupancy(iv).
## Step1. Import the pandas library.
## Step2. Import the dataset.
## Step3. Create two dataframes: one to store the dependent and the other to store independent variables. Suppose df is the original dataframe.

```python
import pandas as pd
```

```python
df=pd.read_csv('F:/Hotel.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
import statsmodels.api as sm
```

```python
X=df['PercentOccupancy'] # The dataframe X is storing the independent variable which is occupancy.
Y=df['Revenueindollar1000'] # The dataframe Y which is storing the dependent variable which in our case in revenue.
```

##### To the X dataframe ,we need to add a constant for making a linear model.

```python
X=sm.add_constant(X) # add_constant is the function which is available in statsmodels library.
```

# To Create the SLR regression model,the python code is given below.

```python
model=sm.OLS(Y,X).fit() #OLS is also a function in statsmodels
model.summary() #To see the summary of the model.
```

# Y(Predicted Variable)=Bo+B1X1 where Y is dv and X is Iv.
### using the model summary,the equation for predicting revenue is:
## PredictedRevenue=(B0)+B1*occupancy
### The value of B0 and B1 can be obtained from the model summary.
### B0=const coef value;B1=PercentOccupancy coeff value
#### PredictedRevenue=(-60.3)+8.2*Occupancy;occupancy=20%,40%,51.2%,etc.
#### From the above equation we can predict what will be the revenue for any particular occupancy.

# How to find the accuracy of the model:-
## To calculate the error in the model: Smaller the error,more accurate the model.
### Create a new variable in python pre_revenue which will store the predicted revenue for every case in the dataset.There should be 12 predicted revenues.
### create a new variable error=Actual reveue-Pre_revenue(12 values)
### Create a new variable rel_error=Error/original revenue(12 values)
### calculate the mean of these 12 values.
### Since there are some negative values ,use the abs() fuction to convert it into positive.

```python
df['pre_revenue']=(-60.3)+(8.2317*df['PercentOccupancy'])
```

```python
df.head()
```

```python
df['error']=abs(df['Revenueindollar1000']-df['pre_revenue'])
```

```python
df.head()
```

```python
df['rel_error']=abs(df['error']/df['Revenueindollar1000'])
```

```python
df.head()
```

```python
df.describe()
```

```python
df.rel_error.mean()*100
```

## The error is roughly 3.5 percent.Therefore the model is highly accurate in predicting revenue from occupancy.

# Use the file hsb2.csv and create a linear regression model to predict read from the variable math and find the accuracy of the model.

```python
X1=df1['math'] # The dataframe X1 is storing the independent variable which is math.
Y1=df1['read'] # The dataframe Y1 is storing the dependent variable which is read.
```

model=sm.OLS(Y1,X1).fit() #OLS is also a function in statsmodels
